```python
import matplotlib.pyplot as plt
```

```python
# Make plots appear and be stored within the notebook
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL 
plt.plot(x, y, 'g')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title("Title")
plt.show()
```

```python
# Subplots
plt.subplot(1, 2, 1)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Red table")
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Blue table")
plt.plot(y, x, 'b')
plt.show()
```

```python
# Object oriented method for creating plots
# Recreating the plot above
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'r')
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title of graph")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes2.plot(y, x)

axes1.set_title("Larger plot")
axes2.set_title("Smaller plot")
```
